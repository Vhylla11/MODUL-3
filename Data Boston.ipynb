```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = load_boston()
dir ( data) ## lihat tabel apa aja pake 'dir'
# data['DESCR'] # lihat deskripsinya ada apa aja
# data['data'] # lihat data
# data [ 'feature_names'] # nama kolom
# data ['filename'] ## lokasinya ada dimana
# data['target'] # y
```

## Buat DataFrame

```python
dfx = pd.DataFrame (data['data'],columns = data['feature_names'])

# df.tail() ## datanya ada 506 kalau yg muncul 505 berrti indexnya 505

dfy = pd.DataFrame (data['target'],columns = ['harga'])
```

## Korelasi terhadap harga

```python
from sklearn import preprocessing
dfxS =pd.DataFrame (preprocessing.scale(dfx),columns=data['feature_names'])

dfxS ['harga'] = dfy
dfxS.corr()
```

# Splitting 90% trainning, 10% testing

```python
x = dfxS [['RM','PTRATIO','LSTAT']]
y = dfxS ['harga']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split (x,y,test_size=0.1) ## untuk fiting model pakai x_train dan y_train

print(x_train.iloc[0])
print (y_train[x_train.iloc[0].name])
```

```python
dfxS.loc[132] ## cek sama atau ga harganya
```

## Membuat model & Prediksi

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

### RM,PTRATIO,LSTAT

```python
model.predict([[0.124466,1.268938,-0.214895]])
```

### Prediksi dari model yang sudah ada yaitu harga 21.4358 , namun data asli = 23.0000

```python
# prediksi untuk x_test index pertama
print(model.predict([x_test.iloc[0]]))

# bandingkan dengan nilai y_test index pertama
print(y_test[x_test.iloc[0].name])
```

### yang diprediksi harganya 12.3731 , harga asli : 7.2

```python
## PREDIKSI SEMUA INDEX

print (model.predict(x_test))
print(y_test)
```

```python
hargaPred = model.predict(x_test)
harga = y_test
```

## HITUNG EVALUATION METRICS


```python
from sklearn.metrics import mean_squared_error
MSE =  mean_squared_error (harga,hargaPred)
MSE
RMSE = np.sqrt(MSE)
RMSE
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(harga,hargaPred)
r2
```

```python
from sklearn.datasets import load_iris
```

```python
i = load_iris ()
dir (i)
# # i['data']
# # i['feature_names']
df = pd.DataFrame (i['data'],columns = ['SL','SW','PL','PW'])
df['spesies'] = i ['target']
df['namaSp'] = df['spesies'].apply (
        lambda x : i['target_names'][x]
    )
df
```

```python
import matplotlib.pyplot as plt

df0 = df[df['spesies'] == 0]
df1 = df[df['spesies'] == 1]
df2 = df[df['spesies'] == 2]

plt.figure(figsize =(14,8))
plt.subplot(121)
plt.plot(df0['SL'],df0['SW'], 'ro')
plt.plot(df1['SL'],df1['SW'], 'go')
plt.plot(df2['SL'],df2['SW'], 'yo')
plt.legend(['setosa','versicolor','virginica'])
plt.title ('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.grid(True)

plt.subplot(122)
plt.plot(df0['PL'],df0['PW'], 'ro')
plt.plot(df1['PL'],df1['PW'], 'go')
plt.plot(df2['PL'],df2['PW'], 'yo')
plt.legend(['setosa','versicolor','virginica'])
plt.title ('Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.grid(True)
```
